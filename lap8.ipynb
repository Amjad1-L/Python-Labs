```python
import os  # استيراد مكتبة os للتعامل مع الملفات والمجلدات

# إنشاء مجلد باسم lab_data إذا ما كان موجود
if not os.path.exists("lab_data"):  # يتحقق إذا المجلد غير موجود
    os.mkdir("lab_data")  # ينشئ المجلد
    print("Folder created")  # يطبع رسالة
else:
    print("Folder already exists")  # إذا موجود يطبع رسالة

# إنشاء مجلدات داخل lab_data
os.makedirs("lab_data/input", exist_ok=True)  # ينشئ input (ولو موجود ما يعطي خطأ)
os.makedirs("lab_data/output", exist_ok=True)  # ينشئ output
os.makedirs("lab_data/logs", exist_ok=True)  # ينشئ logs

# عرض كل العناصر داخل lab_data وتحديد نوعها
items = os.listdir("lab_data")  # يجلب كل الملفات والمجلدات داخل lab_data
for item in items:  # يمر على كل عنصر
    path = os.path.join("lab_data", item)  # يحدد المسار الكامل
    if os.path.isfile(path):  # يتحقق إذا العنصر ملف
        print(item, "is a file")  # يطبع أنه ملف
    else:
        print(item, "is a folder")  # يطبع أنه مجلد

# إعادة تسمية logs إلى history
os.rename("lab_data/logs", "lab_data/history")  # يغير اسم المجلد
print("logs renamed to history")  # يطبع تأكيد
```

```python
# إنشاء ملف notes.txt وكتابة 5 أسطر
with open("lab_data/input/notes.txt", "w") as file:  # يفتح الملف بوضع الكتابة
    file.write("Line 1\n")  # يكتب سطر
    file.write("Line 2\n")  # يكتب سطر
    file.write("Line 3\n")  # يكتب سطر
    file.write("Line 4\n")  # يكتب سطر
    file.write("Line 5\n")  # يكتب سطر

# قراءة الملف وحساب عدد الأسطر
with open("lab_data/input/notes.txt", "r") as file:  # يفتح الملف للقراءة
    lines = file.readlines()  # يقرأ كل الأسطر ويخزنها
    print("Number of lines:", len(lines))  # يطبع عدد الأسطر

# نسخ المحتوى إلى summary.txt
with open("lab_data/input/notes.txt", "r") as source:  # فتح الملف الأصلي
    content = source.read()  # قراءة المحتوى

with open("lab_data/output/summary.txt", "w") as target:  # فتح الملف الجديد
    target.write(content)  # كتابة نفس المحتوى فيه
```
