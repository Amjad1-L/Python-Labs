```python
numbers = []

for i in range(5):
    num = float(input("Enter a number: "))
    numbers.append(num)

total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
words = []

for i in range(5):
    word = input("Enter a word: ")
    words.append(word)

words_tuple = tuple(words)

print("Tuple:", words_tuple)
print("First word:", words_tuple[0])
print("Last word:", words_tuple[-1])
print("Total number of words:", len(words_tuple))

search = input("Enter a word to search: ")

if search in words_tuple:
    print("The word is in the tuple")
else:
    print("The word is NOT in the tuple")
```

```python
students = {}

for i in range(3):
    name = input("Enter student name: ")
    score = float(input("Enter student score: "))
    students[name] = score

print("Students and their scores:")
for name, score in students.items():
    print(name, ":", score)

highest_student = max(students, key=students.get)

print("Student with highest score:", highest_student)
print("Highest score:", students[highest_student])
```

```python
list1 = input("Enter numbers for list 1 separated by space: ").split()
list2 = input("Enter numbers for list 2 separated by space: ").split()

set1 = set(map(int, list1))
set2 = set(map(int, list2))

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference:", set1 - set2)
```

```python
names = []

for i in range(5):
    name = input("Enter a name: ")
    names.append(name)

names_tuple = tuple(names)

print("Tuple:", names_tuple)

name_length_dict = {}

for name in names:
    name_length_dict[name] = len(name)

print("Dictionary:", name_length_dict)

length_set = set(name_length_dict.values())

print("Set of unique name lengths:", length_set)
```
