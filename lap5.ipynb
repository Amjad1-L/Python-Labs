```python
import re
import math


text = input("Enter a paragraph: ")


vowel_words = re.findall(r'\b[AEIOUaeiou]\w*', text)


def count_vowels(paragraph):
    vowels = re.findall(r'[AEIOUaeiou]', paragraph)
    return len(vowels)

total_vowels = count_vowels(text)


lengths = list(map(lambda word: len(word), vowel_words))


if len(lengths) > 0:
    average_length = round(math.fsum(lengths) / len(lengths), 2)
else:
    average_length = 0


print("Words starting with vowels:", vowel_words)
print("Total vowels:", total_vowels)
print("Average vowel-word length:", average_length)
```
