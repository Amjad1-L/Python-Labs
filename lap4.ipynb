```python
total = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    if num % 2 == 0:
        total += num   

print("Final sum =", total)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):

    if i % 3 == 0:
        continue

    product = num * i

    if product > 50:
        break

    print(f"{num} x {i} = {product}")
```

```python
secret = 15   # secret number between 1 and 20
attempts = 0

while True:
    guess = int(input("Guess the number (1-20): "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1   

    if guess == secret:
        print("Correct! 🎉")
        break
    else:
        print("Wrong, try again.")

print("Valid attempts:", attempts)
```

```python
n = int(input("Enter a number n: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count =", count)
```
